```python
# Plotting multiple lines


years = [2000,2001,2002,2003,2004]
apples = [0.3,0.2,0.4,0.3,0.6]
oranges = [0.4,0.6,0.4,0.6,0.8]

plt.plot(years,apples)
plt.plot(years,oranges)

plt.legend(['Apples','Oranges'])
plt.title("Yield")
```

```python
# Histograms
```

```python
df = sns.load_dataset('iris')
```

```python
df.sepal_width
```

```python
plt.title('Distribution of Sepal width')
plt.hist(df.sepal_length)
```

```python
# Controlling the size and number of bins
```

```python
# bins is the interval
import numpy as np
```

```python
plt.hist(df.sepal_width,bins=np.arange(2,5,0.25))
```

```python
# Multiple histograms
```

```python
sentosa_df = df[df.species == 'setosa']
versicolor_df = df[df.species == 'versicolor']
```

```python
sentosa_df
```

```python
plt.hist(sentosa_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25))
plt.hist(versicolor_df.sepal_width,alpha=0.4,bins=np.arange(2,5,0.25))
plt.legend(['Sentosa','versicolor'])
```

```python
plt.hist([sentosa_df.sepal_width,
          versicolor_df.sepal_width],
          bins=np.arange(2,5,0.25),
          stacked=True)

plt.legend(['Sentosa','Versicolor'])
```
